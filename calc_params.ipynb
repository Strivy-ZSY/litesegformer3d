### 用于测试可训练参数量

```python
import torch
import sys
import os
from io import StringIO

from torchsummary import summary

def evaluate_tumor_params():
    """使用torchsummary评估LiteSegFormer3D在Tumor数据集上的可训练参数量"""
    
    in_channels = 4
    num_classes = 3
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    from lsf3d.network_architecture.litesegformer3d_tumor import LiteSegFormer3D
    
    model = LiteSegFormer3D(
        in_channels=in_channels,
        num_classes=num_classes,
        embed_dims=[32, 64, 128, 256],
        patch_kernel_size=[3, 3, 3, 3],
        patch_stride=[2, 2, 2, 2],
        patch_padding=[1, 1, 1, 1],
        depths=[2, 2, 2, 2],
        num_heads=[1, 2, 4, 8],
        sr_ratios=[8, 4, 2, 1],
        decoder_head_embedding_dim=256,
        decoder_dropout=0.0,
        do_ds=False
    )
    
    if device == 'cuda':
        model = model.cuda()
    
    # 捕获torchsummary输出
    old_stdout = sys.stdout
    sys.stdout = mystdout = StringIO()
    
    summary(model, input_size=(in_channels, 128, 128, 128), device=device)
    
    sys.stdout = old_stdout
    output = mystdout.getvalue()
    
    # 仅输出主要信息
    lines = output.strip().split('\n')
    for line in lines[-4:-1]:
        print(line)
```

```python
if __name__ == "__main__":
    evaluate_tumor_params()
```
